# Data Preprocesssing 

```python
# For making content based recommendation system we need some specific features.
# genres, id, keywords, title, overview, cast, crew
movies = movies[['id', 'title', 'keywords', 'genres', 'overview', 'cast', 'crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace= True)
```

```python
movies.isnull().sum()
```

```python
movies.shape
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
# it is coming in a string format
```

```python
import ast
ast.literal_eval(movies.iloc[0].genres)
```

```python
def convert(obj):
    list=[]
    for i in ast.literal_eval(obj):
        list.append(i['name'])
    return list
```

```python
convert(movies.iloc[0].genres)
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
ast.literal_eval(movies['cast'].iloc[0])[0]['name']
```

```python
def convert3(text):
    list=[]
    count = 0
    for i in ast.literal_eval(text):
        if count < 3:
            list.append(i['name'])
        count += 1
    return list
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
ast.literal_eval(movies['crew'].iloc[0])
```

```python
def fetch_director(text):
    list=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            list.append(i['name'])
    
    return list
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
def collapse(list):
    list1 = []
    for i in list:
        list1.append(i.replace(' ',''))
    
    return list1
```

```python
movies['keywords'] = movies['keywords'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['keywords'] + movies['genres'] + movies['overview'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies = movies.drop(['keywords', 'genres', 'overview', 'cast', 'crew'], axis=1)
```

```python
new_movies.head()
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: ' '.join(x))
```

```python
new_movies.head()
```

# Model Training

```python
import nltk
nltk.download('wordnet')
nltk.download('all')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def lem(text):
    list = []
    for i in text.split():
        list.append(lemmatizer.lemmatize(i))
    
    return ' '.join(list)
```

```python
new_movies['tags'].apply(lem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
vectors
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
similarity[0]
```

```python
list(enumerate(similarity[0]))
```

```python
new_movies[new_movies['title'] == 'Avatar'].index[0]
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key =lambda x: x[1])
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key =lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key =lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
```

```python
recommend('Iron Man')
```
